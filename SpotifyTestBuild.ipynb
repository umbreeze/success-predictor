```python
searches=[]
tracks=billboardTracks["title"].tolist()
artists=billboardTracks["artist"].tolist()
```

```python
for k in range(len(tracks)):
  artist_name = artists[k]

  if(artist_name.find(' Featuring ') >= 0):
    artist_name = artist_name.partition(' Featuring ')[0]

  if(artist_name.find(',') >= 0):
    artist_name = artist_name.partition(',')[0]

  search_result = spotifyObject.search(q='track:'+tracks[k]+' '+'artist:'+artist_name+'*' , type='track',limit=1,offset=0,market=None)

  if(search_result['tracks']['total'] == 0):
    search_result = spotifyObject.search(tracks[k], type='track',limit=1,offset=0,market=None)

  searches.append(json.dumps(search_result,sort_keys=True,indent=4))

  if(k % 200 == 0):
    print("{} tracks processed".format(k))
```

```python
print(searches[0])
```

```python
searches_list=[]
for search in searches:
  searches_list.append(json.loads(search))
```
